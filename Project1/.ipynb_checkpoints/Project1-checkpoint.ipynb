```python
import pandas as pd # Импортируем библиотеку для работы с данными
dataset = pd.read_csv('C:/Users/user/ramen-ratings.csv') 
dataset["Review #"]=pd.to_numeric(dataset["Review #"]) # преобразуем сданные в числовой тип
dataset["Stars"]=dataset["Stars"].str.replace("Unrated", "0") # заменяем значения "Unrated" в столбце "Stars" на 0 и преобразуем столбец в числовой тип
dataset["Stars"]=pd.to_numeric(dataset["Stars"]) 
dataset["Style"]=dataset["Style"].str.replace("Cup", "1") # преобразуем строковые значения в числовые
dataset["Style"]=dataset["Style"].str.replace("Bowl", "2")
dataset["Style"]=dataset["Style"].str.replace("Pack", "3") 
dataset["Style"]=dataset["Style"].str.replace("Tray", "4") 
dataset["Style"]=dataset["Style"].str.replace("Box", "5") 
dataset["Style"]=dataset["Style"].str.replace("Can", "6") 
dataset["Style"]=dataset["Style"].str.replace("Bar", "7") 
dataset["Style"]=pd.to_numeric(dataset["Style"])
dataset=dataset.sort_values(by='Review #', ascending=True) # Сортируем данные по столбцу "Review #" по возрастанию
dataset_new = dataset[["Review #", "Brand", "Variety", "Style", "Stars","Country"]]
dataset=dataset.drop("Top Ten", axis=1) # Удаляем незначимые данные, аргумент axis = 1 означает, что данные удаляются по вертикали
dataset ["Review #"] = pd.factorize(dataset["Review #"])[0]
dataset_new.to_csv ("Ramen") # сохраняем обработанный набор данных в новый файл
dataset_new.to_excel ("newramen-ratings.xlsx",  sheet_name='Ramen', index=False, )
dataset_new
```

```python

import pandas as pd # Импортируем библиотеку для работы с данными
import seaborn as sns # Импортируем библиотеку для визуализации данных
dataset = pd.read_csv('C:/Users/user/ramen-ratings.csv')
dataset["Style"]=dataset["Style"].str.replace("Cup", "1")
dataset["Style"]=dataset["Style"].str.replace("Bowl", "2")
dataset["Style"]=dataset["Style"].str.replace("Pack", "3") 
dataset["Style"]=dataset["Style"].str.replace("Tray", "4") 
dataset["Style"]=dataset["Style"].str.replace("Box", "5") 
dataset["Style"]=dataset["Style"].str.replace("Can", "6") 
dataset["Style"]=dataset["Style"].str.replace("Bar", "7") 
dataset["Style"]=pd.to_numeric(dataset["Style"])
dataset["Stars"]=dataset["Stars"].str.replace("Unrated", "0")
dataset["Stars"]=pd.to_numeric(dataset["Stars"])
dataset=dataset.sort_values(by='Review #', ascending=True)
dataset=dataset.drop("Top Ten", axis=1)
dataset = dataset.loc[dataset ['Stars'] < 3]
dataset = dataset.loc[dataset ['Country']=='Taiwan']
data_visualisation = sns.barplot(data=dataset, x="Style", y="Stars")
dataset
```
